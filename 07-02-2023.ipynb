```python
d = pd.read_csv('Windspeed.csv')
d.head(100)
```

```python
d.info
```

```python
d.isnull().sum()
```

# Filing a common Value to All missing data

```python
d.fillna(0)
```

## Adding Missing data to individual values in columns

```python
d.fillna({
    "Temperature":0,
    "Windspeed":5,
    "Status":"sunny"
    
    
}
)
```

## Forward Fill(row)

```python
d.fillna(method='ffill')
```

### Limiting the forward/backward fill

```python
d.fillna(method='ffill',limit=6)
```

### Filling with Pandas Objects 

###### There Are Many Pandas Objects Like df.sum(),df.max(),etc. we can fill the missing value with these two

```python
d.fillna(d.mean())
```

## Filing For Specific range Columns

#### We Can Do Filing for a specific Range of columns too as 


```python
d.fillna(d.mean()['Temperature':'Windspeed' ])
```
